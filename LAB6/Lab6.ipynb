```python
import random

def f(x):
    return -x**2 + 6*x

def get_neighbors(x):
    neighbors = []
    if x - 1 >= 0:
        neighbors.append(x - 1)
    if x + 1 <= 6:
        neighbors.append(x + 1)
    return neighbors

def hill_climbing():
    x = random.randint(0, 6)
    print(f"Initial x = {x}, f({x}) = {f(x)}")
  

    step = 1
    while True:
        current_value = f(x)
        neighbors = get_neighbors(x)

        # Find best neighbor
        best_neighbor = None
        best_value = current_value

        for neighbor in neighbors:
            if f(neighbor) > best_value:
                best_value = f(neighbor)
                best_neighbor = neighbor

        if best_neighbor is None:
            print(f"\nNo better neighbor found... hill climbing stopped.")
            print(f"Final optimal x = {x}, f({x}) = {current_value}")
            break

        print(f"Step {step}: x={x}, current f(x) is {current_value}... move to x={best_neighbor}, f(x)={best_value}")
        x = best_neighbor
        step += 1

hill_climbing()

```

```python
import heapq

def heuristic(n):
    return abs(20 - n)

def get_successors(x):
    return [x + 2, x + 3, x * 2]

def beam_search_task2(start, goal, beam_width=2):

    beam = [(start, [start])]
    level = 0

    while beam:
        print(f"Level {level}: {[state[0] for state in beam]}")

        # Check if goal is in beam
        for number, path in beam:
            if number == goal:
                print("Goal found")
                print("final path:", path)
                return

        # Generate all candidates
        candidates = []
        for number, cur_path in beam:
            for next_num in get_successors(number):
                new_path = cur_path + [next_num]
                candidates.append((next_num, new_path))

        # Keep top k based on heuristic (lowest h value)
        beam = heapq.nsmallest(beam_width, candidates, key=lambda x: heuristic(x[0]))
        level += 1

    print("Goal not reached.")
    return None

beam_search_task2(start=1, goal=20, beam_width=2)

```

```python


def decode(chromosome):
    binary_string = ""
    for bit in chromosome:
        binary_string = binary_string + str(bit)
    return int(binary_string, 2)

def calculate_fitness(chromosome):
    x = decode(chromosome)
    return x**2 + 2*x

def create_random_population():
    chromosome = []
    for i in range(5):
        chromosome.append(random.randint(0, 1))
    return chromosome


def select_parents(population, fitness_scores):
    paired = []
    for i in range(len(population)):
        paired.append((fitness_scores[i], population[i]))
    
    paired.sort(reverse=True)
    
    parents = []
    for i in range(len(paired) // 2):
        parents.append(paired[i][1])
    
    return parents


def crossover(parent1, parent2):
    point = random.randint(1, 4)
    child = parent1[:point] + parent2[point:]
    return child

def mutate(chromosome):
    idx = random.randint(0, 4)
    chromosome[idx] = 1 - chromosome[idx] 
    return chromosome

def genetic_algorithm():

    population_size = 6
    mutation_rate = 0.1
    generations = 15

    # Initialize population
    population = [create_random_population() for _ in range(population_size)]

    best_chromosome = None
    best_x = None
    best_fitness = -1

    for gen in range(generations):
        # Evaluate fitness
        fitness_scores = [calculate_fitness(ind) for ind in population]

        # Track best
        max_fitness = max(fitness_scores)
        best_idx = fitness_scores.index(max_fitness)

        if max_fitness > best_fitness:
            best_fitness = max_fitness
            best_chromosome = population[best_idx][:]
            best_x = decode(population[best_idx])

        print(f"Generation {gen+1}: Best x={decode(population[best_idx])}, Chromosome={population[best_idx]}, Fitness={max_fitness}")

        # Selection
        parents = select_parents(population, fitness_scores)

        # Crossover
        new_population = []
        for _ in range(population_size):
            parent1, parent2 = random.sample(parents, 2)
            child = crossover(parent1, parent2)
            new_population.append(child)

        # Mutation
        for i in range(len(new_population)):
            if random.random() < mutation_rate:
                new_population[i] = mutate(new_population[i])

        population = new_population

    print("final...")
    print(f"Best Chromosome : {best_chromosome}")
    print(f"Best x value    : {best_x}")
    print(f"Best fitness    : {best_fitness}")

genetic_algorithm()
```
